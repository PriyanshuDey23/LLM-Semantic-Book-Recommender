```python
# Description
import numpy as np

books["missing_description"] = np.where(books['description'].isna() , 1,0) # Missing 1 , not missing 0

books['age_of_book']= Current_year - books['published_year']


```

```python
# Interested columns
columns_of_interest = ["num_pages", "age_of_book", "missing_description", "average_rating"]

correlation_matrix = books[columns_of_interest].corr(method = "spearman")

sns.set_theme(style="white")
plt.figure(figsize=(8, 6))
heatmap = sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm",
                      cbar_kws={"label": "Spearman correlation"})
heatmap.set_title("Correlation heatmap")
plt.show()
# Check missing description
# not very strong correlation between missing description and other columns
# closer to 0 , weak correlation
# Closer to 1 or -1 , strong correlation

```

```python
filtered_books=books[(books["description"].isna()) |
      (books["num_pages"].isna()) |
      (books["average_rating"].isna()) |
      (books["published_year"].isna()) ]
# | - or
# in the above  columns , if there is missing description ,num_pages,average_rating,published_year,(Either any of them )
# then we will remove the the rows which contains missing values
# Output:- 303 rows × 14 columns
```

```python
# Delete the observation with missing values
# books that are going to have missing description, num_pages, average_rating, published_year

book_missing = books[~(books["description"].isna()) &
      ~(books["num_pages"].isna()) &
      ~(books["average_rating"].isna()) &
      ~(books["published_year"].isna())
]
# book_missing:- 6507 rows × 14 columns
# 6810-6507=303 (only 303 observations have missing values), so removed
```

```python
book_missing["categories"].value_counts().reset_index().sort_values("count", ascending=False)
# uneven distribution of data
```

```python
# Now lets look at discription category
# Description should contain meaning full information
# Check the length/words of description column 

# book_missing["words_in_description"] = book_missing["description"].str.split().str.len()
book_missing.loc[:, "words_in_description"] = book_missing["description"].str.split().str.len() # Split and then get the length
```

```python
# How many words a description should have to have a meaning full description
# Book description between the given number of words(Check number of words and choose wisely)
book_missing.loc[book_missing["words_in_description"].between(0, 4), "description"]
```

```python
book_missing.loc[book_missing["words_in_description"].between(25, 34), "description"] # Now the description is quite meaning full
```

```python
book_missing_25_words = book_missing[book_missing["words_in_description"] >= 25]
```

```python
book_missing_25_words.head()
# 5197 rows × 15 columns
```

```python
# Subtitle Field

# We will combine the title and subtitle column 
# book_missing_25_words["title_and_subtitle"] = (
#     np.where(book_missing_25_words["subtitle"].isna(), book_missing_25_words["title"],
#              book_missing_25_words[["title", "subtitle"]].astype(str).agg(": ".join, axis=1))
# )

book_missing_25_words.loc[:, "title_and_subtitle"] = (
    np.where(book_missing_25_words["subtitle"].isna(), book_missing_25_words["title"], # When subtitle is missing, use title
             book_missing_25_words[["title", "subtitle"]].astype(str).agg(": ".join, axis=1)) # When subtitle is present, combine title and subtitle
)
```

```python
# Combine the isbn13 and description columns

book_missing_25_words.loc[:,"tagged_description"] = book_missing_25_words[["isbn13", "description"]].astype(str).agg(" ".join, axis=1)
```
