```python
# Mapping top 12 categories with fiction and non fiction
# Converting the top categories to fiction and non fiction
category_mapping = {'Fiction' : "Fiction",
 'Juvenile Fiction': "Children's Fiction",
 'Biography & Autobiography': "Nonfiction",
 'History': "Nonfiction",
 'Literary Criticism': "Nonfiction",
 'Philosophy': "Nonfiction",
 'Religion': "Nonfiction",
 'Comics & Graphic Novels': "Fiction",
 'Drama': "Fiction",
 'Juvenile Nonfiction': "Children's Nonfiction",
 'Science': "Nonfiction",
 'Poetry': "Fiction"}

books["simple_categories"] = books["categories"].map(category_mapping)
```

```python
from transformers import pipeline

fiction_categories = ["Fiction", "Nonfiction"] # For this model we need to specify the categories

pipe = pipeline("zero-shot-classification",
                model="facebook/bart-large-mnli")
```

```python
#  we want to classify whether it is fiction or not
sequence = books.loc[books["simple_categories"] == "Fiction", "description"].reset_index(drop=True)[0] # Get the description
```

```python
sequence
```

```python
pipe(sequence, fiction_categories) # Useing zero shot we will get whether it fiction or nonfiction
# Just like in the simple_categories section it is fiction , the model has also predicted it is as fiction
```

```python
import numpy as np

max_index = np.argmax(pipe(sequence, fiction_categories)["scores"]) # get the highest score
max_label = pipe(sequence, fiction_categories)["labels"][max_index] # based on the score get thelabel
max_label
```

```python
# Convert the above code to function
def generate_predictions(sequence, categories):
    predictions = pipe(sequence, categories)
    max_index = np.argmax(predictions["scores"]) 
    max_label = predictions["labels"][max_index]
    return max_label
```

```python
# Classifier to predict the label and then check it with actual label (simple categories)
# For fiction
from tqdm import tqdm

actual_cats = []
predicted_cats = []

for i in tqdm(range(0, 300)):   # 300 is the number of books
    sequence = books.loc[books["simple_categories"] == "Fiction", "description"].reset_index(drop=True)[i] # Get the description
    predicted_cats += [generate_predictions(sequence, fiction_categories)] # Get the predicted label based on the description
    actual_cats += ["Fiction"] # Get the actual label
```

```python
# For Nonfiction
for i in tqdm(range(0, 300)):
    sequence = books.loc[books["simple_categories"] == "Nonfiction", "description"].reset_index(drop=True)[i]
    predicted_cats += [generate_predictions(sequence, fiction_categories)]
    actual_cats += ["Nonfiction"]
```

```python
predictions_df = pd.DataFrame({"actual_categories": actual_cats, "predicted_categories": predicted_cats}) # Create a dataframe
```

```python
predictions_df
```

```python
# Check the accuracy(percentage of correct predictions) . Actual vs Predicted
predictions_df["correct_prediction"] = (
    np.where(predictions_df["actual_categories"] == predictions_df["predicted_categories"], 1, 0) # If actual and predicted are same then 1 else 0
)
```

```python
predictions_df["correct_prediction"].sum() / len(predictions_df) # Get the accuracy(Correct)
```

```python
isbns = []
predicted_cats = []

missing_cats = books.loc[books["simple_categories"].isna(), ["isbn13", "description"]].reset_index(drop=True) # Get the missing categories # Description
```

```python
# Predict the missing value (Categories in simple_categories)
for i in tqdm(range(0, len(missing_cats))):
    sequence = missing_cats["description"][i] # Get the description
    predicted_cats += [generate_predictions(sequence, fiction_categories)] # Get the predicted label
    isbns += [missing_cats["isbn13"][i]] # Get the isbn
```

```python
missing_predicted_df = pd.DataFrame({"isbn13": isbns, "predicted_categories": predicted_cats})
```

```python
missing_predicted_df
```

```python
# Merging it into original data frame and then replacing the missing values with the predicted values in simple_categories column
books = pd.merge(books, missing_predicted_df, on="isbn13", how="left") # Merge the missing values with the original dataframe
books["simple_categories"] = np.where(books["simple_categories"].isna(), books["predicted_categories"], books["simple_categories"]) # Replace the missing values with the predicted values
# above:- when the value is na , then replace it with the predicted value else keep the original value
books = books.drop(columns = ["predicted_categories"]) # Drop the predicted values column
```
